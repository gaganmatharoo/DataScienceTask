# Detecting Outliers

```python
import pandas as pd

weather = pd.read_csv('data\weather.csv')
```

```python
from matplotlib import pyplot as plt
```

```python
_windchillm = weather[" _windchillm"].tolist()

plt.hist(_windchillm)
plt.show()
```

```python
lower_bound = 0.1
upper_bound = 0.95

res = weather[' _windchillm'].quantile([lower_bound, upper_bound])
res
```

```python
true_index = (res.loc[lower_bound] < weather[' _windchillm'].values) & (weather[' _windchillm'].values < res.loc[upper_bound])
true_index

```

# Plot of rain, snow, pressure etc

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
weather = pd.read_csv("data\weather.csv", parse_dates=['datetime_utc'],skipinitialspace=True)
```

```python
weather['Date'] = pd.to_datetime(weather['datetime_utc'])
weather['year'] = weather['Date'].dt.year
weather['month'] = weather['Date'].dt.month
weather['day'] = weather['Date'].dt.day
weather['hour'] = weather['Date'].dt.hour
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

yearHumi = weather.groupby(weather.year).mean()
yearHumi._hum.plot(ax=axes[0,0])#HUMIDITY

yearPres = weather.groupby(weather.year).mean()
yearPres._pressurem.plot(ax=axes[0,1])#PRESSURE

yearRain = weather.groupby(weather.year).mean()
yearRain._rain.plot(ax=axes[1,0])#RAIN

yearHeat = weather.groupby(weather.year).mean()
yearHeat._heatindexm.plot(ax=axes[1,1])#HEAT

fig.subplots_adjust(hspace=0.5, wspace=0.5)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

yearDew = weather.groupby(weather.year).mean()
yearDew._dewptm.plot(ax=axes[0,0])
plt.title('Delhi "dewptm" per year')

yearWdird = weather.groupby(weather.year).mean()
yearWdird._wdird.plot(ax=axes[0,1])
plt.title('Delhi "wdird" per year')

yearWspdm = weather.groupby(weather.year).mean()
yearWspdm._wspdm.plot(ax=axes[1,0])
plt.title("Delhi Wind Speed per year")

yearTor = weather.groupby(weather.year).mean()
yearTor._tornado.plot(ax=axes[1,1])
plt.title('Delhi "dewptm" per year')


fig.subplots_adjust(hspace=0.5, wspace=0.5)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

yearSnow = weather.groupby(weather.year).mean()
plt.title('Snow in Delhi per year')
yearSnow._snow.plot(ax=axes[0])

yearThun = weather.groupby(weather.year).mean()
plt.title('Thunder in Delhi per year')
yearThun._thunder.plot(ax=axes[1])

fig.subplots_adjust(hspace=0.5, wspace=0.5)
```
